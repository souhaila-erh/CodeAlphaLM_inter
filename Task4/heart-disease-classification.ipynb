
## ðŸ“ Project Description

The objective of this project is to **predict the presence or absence of heart disease** in a patient based on various medical attributes. This is a **binary classification task**, where the model learns from patient data and determines whether they are likely to suffer from heart disease or not.

## ðŸ“Š Dataset Overview

The dataset used in this project is publicly available on Kaggle:  
ðŸ”— [Heart Disease Diagnosis Dataset](https://www.kaggle.com/datasets/utkarshx27/heart-disease-diagnosis-dataset)

It contains **14 features** collected from clinical tests, including age, sex, blood pressure, cholesterol levels, electrocardiographic results, and more. The final column `heart disease` serves as the **target variable**, indicating whether the patient has heart disease (`1`) or not (`0`).

### ðŸ” Goal

The main goal is to:
- Explore and understand the data through visualization and statistics.
- Build and compare various classification models (e.g., Random Forest, SVM, Logistic Regression, etc.).
- Evaluate model performance using appropriate metrics.
- Derive insights that could help in medical diagnosis or decision-making.

## ðŸ’¡ Why is this important?

Heart disease is a leading cause of death globally. **Early prediction** using machine learning models can support doctors in diagnosing patients more efficiently and taking preventive measures.



## ðŸ§¬ Feature Description

This dataset contains **14 columns**, each representing a medical attribute or indicator related to a patient's heart health.

| No. | Column Name         | Description |
|-----|---------------------|-------------|
| 1   | `age`               | Age of the patient (in years) |
| 2   | `sex`               | Sex of the patient (1 = male, 0 = female) |
| 3   | `cp`                | Chest pain type (4 possible values) |
| 4   | `trestbps`          | Resting blood pressure (in mm Hg) |
| 5   | `chol`              | Serum cholesterol level (in mg/dl) |
| 6   | `fbs`               | Fasting blood sugar > 120 mg/dl (1 = true, 0 = false) |
| 7   | `restecg`           | Resting electrocardiographic results (values: 0, 1, 2) |
| 8   | `thalach`           | Maximum heart rate achieved |
| 9   | `exang`             | Exercise-induced angina (1 = yes, 0 = no) |
| 10  | `oldpeak`           | ST depression induced by exercise relative to rest |
| 11  | `slope`             | Slope of the peak exercise ST segment (0 = downsloping, 1 = flat, 2 = upsloping) |
| 12  | `ca`                | Number of major vessels (0-3) colored by fluoroscopy |
| 13  | `thal`              | Thalassemia: 3 = normal, 6 = fixed defect, 7 = reversible defect |
| 14  | `heart disease`     | Target variable: presence (1) or absence (0) of heart disease |

ðŸ“Œ **Note:** The target column is named **`heart disease`**, where:
- `1` indicates absence of heart disease  
- `2` indicates presence of heart disease

Splitting Data into Features and Target Variable


```python
X = df.drop(columns='heart disease', axis=1)
Y = df['heart disease']
```

### Splitting the Data into Training and Testing Sets

In this step, the data is split into training and testing sets:

- **X_train, y_train**: The training data, which will be used to train the model.
- **X_test, y_test**: The testing data, which will be used to evaluate the model's performance.

- `test_size=0.2`: This means that 20% of the data is reserved for testing, and 80% is used for training.
- `stratify=Y`: This ensures that the distribution of the target variable (`heart disease`) is preserved in both the training and testing sets.
- `random_state=2`: This ensures that the split is reproducible by using a fixed random seed.

This split helps ensure that the model is trained and evaluated on different subsets of the data, which aids in assessing its generalization ability.


```python

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

# Model Training and Evaluation: Comparing Multiple Classifiers

In this analysis, several machine learning models were trained and evaluated to predict the presence or absence of heart disease. The following methodology was followed:


##  Models Selection:
- Six classification models were selected for evaluation:
    - **Random Forest Classifier**: A robust ensemble method for classification that handles imbalanced data well using class weights.
    - **Gradient Boosting Classifier**: An ensemble method that builds decision trees in a sequential manner to minimize the prediction error.
    - **Support Vector Machine (SVM)**: A powerful classifier using hyperplanes for classification, with class weights to handle imbalances.
    - **Logistic Regression**: A linear model used for binary classification, also with class weights.
    - **K-Nearest Neighbors (KNN)**: A simple and intuitive model that classifies data points based on their proximity to others.
    - **Decision Tree Classifier**: A tree-based model that makes decisions based on splitting data features.

##  Class Weight Handling:
- For models like Random Forest, SVM, Logistic Regression, and Decision Tree, the `class_weight='balanced'` parameter was added to address the class imbalance in the target variable (heart disease presence vs. absence).

##  Training and Evaluation:
- Each model was trained on the training data (`X_train` and `y_train`), and predictions were made on the test data (`X_test`).
- The evaluation metrics included:
    - **Accuracy Score**: Measures the percentage of correct predictions.
    - **Classification Report**: Provides detailed metrics such as precision, recall, and F1-score for each class.
    - **Confusion Matrix**: Visualizes the number of correct and incorrect predictions, highlighting the types of errors (false positives and false negatives).

##  Model Selection:
- After evaluating all models, the one with the highest accuracy was selected as the best model.
- The best model was stored for future use.

##  Visualization:
- Confusion matrices were plotted for each model to help visualize the performance and identify areas where the model could be improved.


```python

# Liste des modÃ¨les Ã  tester avec ajout du paramÃ¨tre class_weight
models = [
    ('Random Forest', RandomForestClassifier(random_state=42, class_weight='balanced')),
    ('Gradient Boosting', GradientBoostingClassifier(random_state=42)),
    ('Support Vector Machine', SVC(random_state=42, class_weight='balanced', probability=True)),
    ('Logistic Regression', LogisticRegression(random_state=42, max_iter=1000, class_weight='balanced')),
    ('K-Nearest Neighbors', KNeighborsClassifier()),
    ('Decision Tree', DecisionTreeClassifier(random_state=42, class_weight='balanced'))  # âž• ajoutÃ© ici
]
best_model = None
best_accuracy = 0.0

# EntraÃ®ner et Ã©valuer chaque modÃ¨le
for name, model in models:
    # EntraÃ®nement
    model.fit(X_train, y_train)

    # PrÃ©diction
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]  # ProbabilitÃ©s pour ROC et AUC

    # Ã‰valuation de la prÃ©cision
    accuracy = accuracy_score(y_test, y_pred)

    # Classification Report
    print(f"Model: {name}")
    print(f"Test Accuracy: {accuracy:.4f}")
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    
    # Matrice de Confusion
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["No Disease", "Disease"], yticklabels=["No Disease", "Disease"])
    plt.title(f"Confusion Matrix - {name}")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    
    # Sauvegarde du meilleur modÃ¨le
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = model

# Affichage du meilleur modÃ¨le
print("Best Model:", best_model)
```

# Model Training and Hyperparameter Tuning

After training the models, we performed hyperparameter tuning using **GridSearchCV** for each model. 

##  Grid Search and Evaluation:
- For each model, we performed **GridSearchCV** with cross-validation (5-folds) to search for the best hyperparameters.
- **Accuracy** was used as the scoring metric for tuning the hyperparameters.
- After finding the best hyperparameters, the models were evaluated on the test data (`X_test` and `y_test`).

## Model Selection:
- The best model for each algorithm was selected based on its accuracy on the test data.
- The **best model** overall, with the highest accuracy, was retained for future use.

## Results:
For each model, the best hyperparameters, test accuracy, and classification report (including precision, recall, and F1-score) were displayed.



```python


# Dictionnaire des modÃ¨les avec les hyperparamÃ¨tres Ã  tester
models_with_params = {
    'Random Forest': (RandomForestClassifier(random_state=42, class_weight='balanced'), {  # class_weight ajoutÃ©
        'n_estimators': [50, 100],
        'max_depth': [None, 10, 20]
    }),
    
    'Gradient Boosting': (GradientBoostingClassifier(random_state=42), {
        'n_estimators': [50, 100],
        'learning_rate': [0.01, 0.1],
        'max_depth': [3, 5]
    }),
    
    'Support Vector Machine': (SVC(random_state=42, class_weight='balanced'), {  # class_weight ajoutÃ©
        'C': [0.1, 1, 10],
        'kernel': ['linear', 'rbf']
    }),
    
    'Logistic Regression': (LogisticRegression(random_state=42, max_iter=3000, class_weight='balanced'), {  # class_weight ajoutÃ©
        'C': [0.01, 0.1, 1, 10]
    }),
    
    'K-Nearest Neighbors': (KNeighborsClassifier(), {
        'n_neighbors': [3, 5, 7]
    }),
    
    'Decision Tree': (DecisionTreeClassifier(random_state=42, class_weight='balanced'), {  # class_weight ajoutÃ©
        'max_depth': [None, 5, 10],
        'criterion': ['gini', 'entropy']
    })
}

best_model = None
best_accuracy = 0.0

# Boucle sur chaque modÃ¨le
for name, (model, params) in models_with_params.items():
    print(f"\nTraining model: {name}")
    
    grid = GridSearchCV(model, params, cv=5, scoring='accuracy', n_jobs=-1)
    grid.fit(X_train, y_train)
    
    best_estimator = grid.best_estimator_
    
    # PrÃ©diction sur les donnÃ©es de test
    y_pred = best_estimator.predict(X_test)
    
    # Ã‰valuation
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Best Parameters: {grid.best_params_}")
    print(f"Test Accuracy: {accuracy:.4f}")
    print(classification_report(y_test, y_pred))

    # Mise Ã  jour du meilleur modÃ¨le
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_model = best_estimator

print("\nBest Model Overall:", best_model)
```

After evaluating multiple models and performing hyperparameter tuning, the **RandomForestClassifier** was selected as the best model for predicting heart disease.
